# VGG Pretrained model in keras

Imports

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

Downloading and loading VGG16 model

```python
# load pre trained VGG16 model from keras applications
model = VGG16(weights = 'imagenet')
print(model.summary())
```

```python
from tensorflow.keras.applications import VGG19

# Load the pretrained VGG19 model from keras applications
model = VGG19(weights = 'imagenet')
print("Model Summary:")
print(model.summary())
```

# Prediction using the pretrained model

```python
# Prediction function
def predict_image(model, image_path):
  # Load the model with required target size for VGG16
  img = load_img(image_path, target_size=(224, 224))

  # Convert the image into numpy array
  img_array = img_to_array(img)

  # Add a batch dimension (1, 224, 224, 3)
  img_array = np.expand_dims(img_array, axis=0)

  # Preprocess the image for VGG16
  img_array = preprocess_input(img_array)

  # predict the image class
  predictions = model.predict(img_array)

  # Decode the predictions into readable labels
  decoded_predictions = decode_predictions(predictions, top=5)
  return decoded_predictions[0]
```

Passing images for prediction

```python
# Sample image prediction
sample_image_path = '/content/bus.jpg'

# Predict the image class
predictions = predict_image(model, sample_image_path)

# Display the predictions
print("Peredictions:")
for i, (imagenet_id, label, scores) in enumerate(predictions):
  print(f"{i+1}: {label}: ({scores:.2f})")
```

```python
# Visualize the sample image
img = load_img(sample_image_path, target_size =(224, 224))
plt.imshow(img)
plt.title('Sample Image')
plt.axis('off')
plt.show()
```
